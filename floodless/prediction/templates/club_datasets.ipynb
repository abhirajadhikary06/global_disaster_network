```python
import pandas as pd

# List of CSV file names
csv_files = ["/Users/anikchand/Documents/Floodless/floodless/prediction/templates/predicted_datasets/draught_predicted.csv", "/Users/anikchand/Documents/Floodless/floodless/prediction/templates/predicted_datasets/earthquake_predicted.csv", "/Users/anikchand/Documents/Floodless/floodless/prediction/templates/predicted_datasets/flood_predicted.csv", "/Users/anikchand/Documents/Floodless/floodless/prediction/templates/predicted_datasets/storm_predicted.csv", "/Users/anikchand/Documents/Floodless/floodless/prediction/templates/predicted_datasets/volcano_predicted.csv", "/Users/anikchand/Documents/Floodless/floodless/prediction/templates/predicted_datasets/wildfire_predicted.csv"]

# Read all CSVs and concatenate them row-wise (axis=0)
df_list = [pd.read_csv(file) for file in csv_files]
result = pd.concat(df_list, axis=0, ignore_index=True)  # Stacks them vertically

# Save to a new CSV
result.to_csv("final_dataset.csv", index=False)

print(result.shape)  # Check the final shape
```
