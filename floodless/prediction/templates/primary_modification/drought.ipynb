```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
script_dir = os.path.dirname(os.path.abspath(os.getcwd()))
file_path = os.path.join(script_dir, "raw_datasets", "drought.xlsx")
df = pd.read_excel(file_path)
```

```python
df.head()
```

```python
df.columns
```

```python
df.rename(columns={df.columns[0]: 'Year'}, inplace=True)
```

```python
df.info()
```

```python
df
```

```python
required_columns = ['Year', 'Disaster Type', 'Country', 'Location', 'Total Affected', 'Magnitude']
df = df[required_columns]
```

```python
df
```

```python
df['Year'] = df['Year'].astype(str).str[:4]
```

```python
df['Location'] = df['Location'].str.split(',').str[0]
```

```python
df.shape
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.isnull().mean() * 100
```

```python
sns.violinplot(x = 'Magnitude', data = df)
```

```python
df_min = df['Magnitude'].min()
# df_max = df['Magnitude'].max()

df['Magnitude'] = df['Magnitude'].apply(
    lambda x: round(np.random.uniform(df_min, 300000), 1) if pd.isna(x) else x
)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Location'].fillna(df['Location'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.violinplot(x = 'Total Affected', data = df)
```

```python
df['Total Affected'].fillna(df['Total Affected'].mode()[0], inplace=True)
```

```python
df
```

```python
df.to_csv("drought_cleaned.csv", index=False)
```
